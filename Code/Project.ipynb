### Text Cleaning

```python
import re
import string
import nltk
```

```python
def clean_text(text):
    text = text.lower()                                  # lower-case all characters
#     text =  re.sub(r'@\S+', '',text)                     # remove twitter handles
    text =  re.sub(r'http\S+', '',text)                  # remove urls
    text =  re.sub(r'pic.\S+', '',text) 
    text =  re.sub(r"[^a-zA-Z+']", ' ',text)             # only keeps characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text+' ')      # keep words with length>1 only
    text = "".join([i for i in text if i not in string.punctuation])
    words = nltk.tokenize.word_tokenize(text)
    stopwords = nltk.corpus.stopwords.words('english')   # remove stopwords
    text = " ".join([i for i in words if i not in stopwords and len(i)>2])
    text= re.sub("\s[\s]+", " ",text).strip()            # remove repeated/leading/trailing spaces
    return text
```

```python
data_df['cleaned_description'] = data_df.Description.apply(clean_text)
data_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

x_train, x_test, y_train, y_test = train_test_split(data_df['cleaned_description'], data_df['Category'], test_size=0.2, random_state=42)

vectorizer = CountVectorizer()

x_train_cv = vectorizer.fit_transform(x_train)
x_test_cv = vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

nb_classifier = MultinomialNB()

nb_classifier.fit(x_train_cv, y_train)

y_pred_cv = nb_classifier.predict(x_test_cv)

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred_cv)
precision = precision_score(y_test, y_pred_cv, average='weighted')
recall = recall_score(y_test, y_pred_cv, average='weighted')
f1 = f1_score(y_test, y_pred_cv, average='weighted')

print("Overall Performance Metrics(Unigram):")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Calculate performance metrics for each label
labels = ['Social', 'Entertainment', 'Productivity', 'Developer Tools',
       'Personalization', 'Utilities & Tools', 'Multimedia Design',
       'Business', 'Education', 'Security', 'Books & Reference',
       'Lifestyle', 'Sports', 'Games', 'Educational']

classification_report_result_cv = classification_report(y_test, y_pred_cv, labels=labels)
print("\nPerformance Metrics for Each Label:")
print(classification_report_result_cv)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Create SVM classifier
svm_classifier = SVC(kernel='linear')

# Fit the classifier on the training data
svm_classifier.fit(x_train_cv, y_train)

# Predict on the test data
y_pred_svm = svm_classifier.predict(x_test_cv)

# Calculate performance metrics
accuracy_svm = accuracy_score(y_test, y_pred_svm)
precision_svm = precision_score(y_test, y_pred_svm, average='weighted')
recall_svm = recall_score(y_test, y_pred_svm, average='weighted')
f1_svm = f1_score(y_test, y_pred_svm, average='weighted')

print("Overall Performance Metrics (SVM):")
print("Accuracy:", accuracy_svm)
print("Precision:", precision_svm)
print("Recall:", recall_svm)
print("F1 Score:", f1_svm)

# Calculate performance metrics for each label
labels = ['Social', 'Entertainment', 'Productivity', 'Developer Tools',
       'Personalization', 'Utilities & Tools', 'Multimedia Design',
       'Business', 'Education', 'Security', 'Books & Reference',
       'Lifestyle', 'Sports', 'Games', 'Educational']


# Calculate performance metrics for each label
classification_report_result_svm = classification_report(y_test, y_pred_svm, labels=labels)
print("\nPerformance Metrics for Each Label (SVM):")
print(classification_report_result_svm)
```
