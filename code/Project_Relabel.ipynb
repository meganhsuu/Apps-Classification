```python
def clean_category(category):
    
    # Split the string based on '>'
    parts = category.split('>')
    
    # Extract the first part and remove leading/trailing whitespaces
    cleaned_category = parts[0].strip()
    return cleaned_category

data_df['Category'] = data_df['Category'].apply(clean_category)
```

```python
data_df['Category'].unique()
```

```python
# Define a mapping dictionary for the main categories
category_mapping = {
    'Social': 'Social',
    'Entertainment': 'Entertainment',
    'Music': 'Entertainment',
    'Photo & video': 'Entertainment',
    'Productivity': 'Productivity',
    'Developer tools': 'Developer Tools',
    'Personalization': 'Personalization',
    'Utilities & tools': 'Utilities & Tools',
    'Multimedia design': 'Multimedia Design',
    'Shopping': 'Shopping',
    'Business': 'Business',
    'Education': 'Education',
    'Security': 'Security',
    'Books & reference': 'Books & Reference',
    'Medical': 'Medical',
    'Personal finance': 'Personal Finance',
    'News & weather': 'News & Weather',
    'Kids & family': 'Kids & Family',
    'Health & fitness': 'Health & Fitness',
    'Lifestyle': 'Lifestyle',
    'Sports': 'Sports',
    'Travel': 'Travel',
    'Racing & flying': 'Games',
    'Strategy': 'Games',
    'Action & adventure': 'Games',
    'Family & kids': 'Games',
    'Multi-Player Online Battle Arena': 'Games',
    'Shooter': 'Games',
    'Simulation': 'Games',
    'Card & board': 'Games',
    'Puzzle & trivia': 'Games',
    'Classics': 'Games',
    'Casino': 'Games',
    'Other': 'Other',
    'Educational': 'Educationl',
    'Platformer': 'Games',
    'Fighting': 'Games',
    'Role playing': 'Games',
    'Word': 'Games',
    'Food & dining': 'Food & Dining',
    'Navigation & maps': 'Navigation & Maps',
    'Tools': 'Tools'
}

# Map the categories to main categories using the mapping dictionary
data_df['Main_Category'] = data_df['Category'].map(category_mapping)

# Print the DataFrame with the new 'Main_Category' column
print(data_df[['Category', 'Main_Category']])
```

```python
print(data_df['Main_Category'].value_counts())
```

```python
# Get the names of the top 15 categories
top_15_categories_names = data_df['Main_Category'].value_counts().head(15).index

# Filter the DataFrame to include only categories in the top 15
filtered_data_df = data_df[data_df['Main_Category'].isin(top_15_categories_names)]

# Print the filtered DataFrame
filtered_data_df
```

```python
# Define the file path for saving the CSV file
file_path = "windows_store_top15_data.csv"

# Save the filtered DataFrame to a CSV file
filtered_data_df.to_csv(file_path, index=False)
```
